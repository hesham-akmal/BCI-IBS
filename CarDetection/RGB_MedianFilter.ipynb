```python
import numpy as np , math
from skimage import io
%matplotlib inline
```

```python
def RGB_MedianFilter(img , mask_size):
    R,C = img.shape[0] , img.shape[1]
    R_img = img[:,:,0]
    G_img = img[:,:,1]
    B_img = img[:,:,2]
    t = mask_size//2
    u = t+1
    
    fltrd_img = np.zeros(img.shape, dtype = img.dtype)
    fltrd_img_R = fltrd_img[:,:,0]
    fltrd_img_G = fltrd_img[:,:,1]
    fltrd_img_B = fltrd_img[:,:,2]
    
    for i in range(t,R-t):
        for j in range(t,C-t): 
            R_window = R_img[i-t:i+u , j-t:j+u]
            fltrd_img_R[i,j] = np.median(R_window)
            
            G_window = G_img[i-t:i+u , j-t:j+u]
            fltrd_img_G[i,j] = np.median(G_window)
            
            B_window = B_img[i-t:i+u , j-t:j+u]
            fltrd_img_B[i,j] = np.median(B_window)
            
    fltrd_img[:,:,0] = fltrd_img_R
    fltrd_img[:,:,1] = fltrd_img_G
    fltrd_img[:,:,2] = fltrd_img_B
    
            
    return fltrd_img   
```

```python
img = io.imread('test0.jpg')
io.imshow(img)
print(img.dtype)
```

```python
img2 = RGB_MedianFilter(img,7)
```
